```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("employees.csv")
df.head()
```

# Scatter plot

```python
sns.relplot(x=df["Experience"], y=df["Salary"])
plt.yticks(np.linspace(5000, 40000, 8), [f"{value:.0f}k" for value in np.linspace(5, 40, 8)])
plt.show()
```

In above scatter plot, while the points are plotted in two dimensions, another dimension can be added to the plot by coloring the points according to a third variable. 

```python
sns.relplot(x=df["Experience"], y=df["Salary"], hue=df["Department"], alpha=0.8)
plt.yticks(np.linspace(5000, 40000, 8), [f"{value:.0f}k" for value in np.linspace(5, 40, 8)])
plt.show()
```

Note that we can also use the size parameter as a hue-like variable, where the third variable is differentiated based on the size of the markers. Additionally, the sizes parameter allows us to control the range of marker sizes for better visualization.

```python
sns.relplot(x=df["Experience"], y=df["Salary"], size=df["Satisfaction"], sizes=(5, 80), alpha=0.7)
plt.yticks(np.linspace(5000, 40000, 8), [f"{value:.0f}k" for value in np.linspace(5, 40, 8)])
plt.show()
```

# Line plot

```python
import warnings
warnings.filterwarnings("ignore")
```

To demonstrate line plot, we load the built-in data set from seaborn library.

```python
data = sns.load_dataset("dowjones")
data.head()
```

```python
sns.relplot(x=data["Date"], y=data["Price"], kind="line")
plt.show()
```

More complex datasets will have multiple measurements for the same value of the x variable. The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the 95% confidence interval around the mean. This is not possible with simple plt.plot() function. 

```python
print("Consider the built-in dataset given below")
data= sns.load_dataset("fmri")
data.head()
```

```python
print("If you want to ignore this error bar, use 'errorbar=None'")
sns.relplot(x=data["timepoint"], y=data["signal"], kind="line")
plt.grid()
plt.show()
```

```python
print("If we want to represent the spread of the distribution at each point by plotting the standard deviation instead of a confidence interval, we can use errorbar='sd'.")
```

```python
sns.relplot(x=data["timepoint"], y=data["signal"], kind="line", errorbar="sd")
plt.grid()
plt.show()
```

```python
sns.relplot(x=data["timepoint"], y=data["signal"], hue=data["event"],kind="line", errorbar="sd")
plt.grid()
plt.show()
```

```python
print("Note that 'size' and 'style' parameter can be also used in line plot as hue-like variables.")
```

To see the multiple relationships between the variables, we can make multiple axes and plot subsets of the data on each of them. This is applicable for both kind of plots, line and scatter.

```python
data = sns.load_dataset("tips")
data.head()
```

```python
sns.relplot(x=data["total_bill"], y=data["tip"], col=data["time"], height=3)
plt.show()
```

```python
print("In above line, we use 'col' parameter to see relationships between multiple columns. 'col' split the graph in multiple columns. we can also use 'row' parameter which split the graph in multiple rows.")
```

```python
sns.relplot(x=data["total_bill"], y=data["tip"], col=data["time"], row=data["sex"], height=3)
plt.show()
```
